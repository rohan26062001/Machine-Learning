# Step 1 : Importing Libraries and Laoding the Dataset

```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Loading data
df = pd.read_csv('insurance.csv')
print(df.shape)
```

# Step 2 : Calculating 'm' and 'c' for Regresssion Line

```python
# Defining Variables
X = df['age'].values
Y = df['bmi'].values

# Calculating Size
n = len(X)

# Calculating Mean of x and y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Calculating 'm' and 'c'
num = 0
den = 0
for i in range (n):
    num += (X[i] - mean_x)*(Y[i] - mean_y)
    den += (X[i] - mean_x)**2

m = num/den
c = mean_y - (m*mean_x)

print(m,c)
```

# Step 3 : Plotting Graph

```python
# Plotting Graph
min_x = np.min(X)
max_x = np.max(Y) + 10

# For Line
x = np.linspace(min_x, max_x, 1000)
y = m*x + c
plt.plot(x,y,'r-',label="Regression Line", linewidth=3)

# For Other Data Points
plt.scatter(X,Y,c='g',label="Age/BMI Points")

plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend()
plt.show()
```

# Step 4 : Calculating R^2

```python
ss_r = 0
ss_t = 0

for i in range(n):
    y_pred = m*X[i]+c
    ss_r += (Y[i] - y_pred)**2
    ss_t += (Y[i] - mean_y)**2

r2 = 1 - (ss_r/ss_t)
print(r2)
```

### Conclusion - No Relation Between Age and BMI
