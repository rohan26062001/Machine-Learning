# Matplotlib Tutorial

```python
# Includng Libraries
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plotting to our Canvas
x = [1,2,3]
y = [4,5,1]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('First Graph')
plt.plot(x,y)
```

```python
# Including Style from Matplotlib
from matplotlib import style
style.use("ggplot")
```

```python
# Plotting to Our Canvas
x1 = [1,2,3]
y1 = [4,5,1]

x2 = [8,3,6]
y2 = [45, 5, 89]

# Plotting
plt.plot(x1,y1,'g',label="First Line", linewidth=2)
plt.plot(x2,y2,'c',label="Second Line", linewidth=3)

# Adding Legend
plt.legend()

# Giving Title
plt.title('Second Graph')

# Adding Grid
plt.grid(True, color = 'k')
```

## Bar Graph

```python
# Plotting Bar Graph
plt.bar([1,3,5,7,9],[5,2,7,8,2],label="Bar Graph One")

plt.bar([2,4,6,8,10],[8,6,2,5,6],label="Bar Graph Two",color="g")

plt.legend()

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.title("Bar Graph")
```

## Histogram

```python
# Plotting Histogram
mu = 121 
sigma = 21
x5 = mu + sigma * np.random.randn(1000)
y5=100

plt.hist(x5,y5,histtype='bar',rwidth=0.8)

plt.legend()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Histogram')
```

## Scatter Plot

```python
x6 = [1,2,3,4,5,6,7,8]
y6 = [5,2,4,2,1,2,5,2]

plt.scatter(x6, y6, label='skitscat', color='g', s=25, marker='o')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.title("Scatter Plot Diagram")
```

## Stack Plot

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color='m',label='Sleeping',linewidth=5)
plt.plot([],[],color='c',label='Eating',linewidth=5)
plt.plot([],[],color='r',label='Working',linewidth=5)
plt.plot([],[],color='k',label='Playing',linewidth=5)

plt.stackplot(days, sleeping, eating, working, playing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Area Plot')
```

## Pie Charts

```python
# Pie Charts
slices = [7,2,3,13]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['c', 'm', 'r', 'b']

plt.pie(slices, labels=activities, colors=cols, startangle=90, shadow=True, explode=(0,0.1,0,0), autopct='%1.1f%%')

plt.title('Pie Chart')
```

## Multiple Plots

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0,0.5,0.1)
t2 = np.arange(0.0,0.5,0.02)

plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2))

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2))
```
